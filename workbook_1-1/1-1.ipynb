```python
from astropy.io import fits
from astropy.table import Table
from pathlib import Path
import numpy as np
```

```python
data_dir = "fits"
lname = "files.lst"
my_path = Path(f'./{data_dir}/{lname}')

# fixing path nonsense since I put the .fits in a dir
fits_files = []

for line in Path(my_path).read_text().splitlines():
    fits_files.append(line[2:])

# reading the hduls into an array
all_data = []
i = 0

for fname in fits_files:
    with fits.open(f'{data_dir}/{fname}') as hdul:
        # print(repr(hdul[2].data))
        all_data.append(hdul[0])
        # finding that damn Z header
        
        if i == 0:
            # print(repr(hdul[2].header))
            # for head in hdul[2].header:
            #     if head == 'TTYPE64':
            #         print(f'I found {hdul[2].header[head]}!')
            print(repr(hdul[0].header))
            i += 1
        

all_data = np.array(all_data)
```
