```python
from astropy.io import fits
from astropy.table import Table
from pathlib import Path
import numpy as np
```

```python
data_dir = "fits"
lname = "files.lst"
my_path = Path(f'./{data_dir}/{lname}')

# fixing path nonsense since I put the .fits in a dir
fits_files = []

for line in Path(my_path).read_text().splitlines():
    fits_files.append(line[2:])

# inputting all the  
all_data = []

i = 0

for fname in fits_files:
    with fits.open(f'{data_dir}/{fname}') as hdul:
        all_data.append(hdul)

all_data = np.array(all_data)
```

```python
rows = []

with fits.open("fits/spec-0283-51959-0350.fits") as hdul:
    print(hdul.info())
    for i, hdu in enumerate(hdul):
        for key, value in hdu.header.items():
            rows.append((i, key, value, hdu.header.comments[key]))

header_table = Table(
    rows=rows,
    names=("HDU", "Keyword", "Value", "Comment")
)

header_table
```
