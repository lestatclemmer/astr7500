```python
from astropy.io import fits
from astropy.table import Table
from pathlib import Path
import numpy as np
```

```python
data_dir = Path("fits")
lst_file = data_dir / "files.lst"

fits_files = [line.strip() for line in lst_file.read_text().splitlines() if line.strip()]


# reading the hduls into an array
all_data = []
i = 0

for fname in fits_files:
    with fits.open(f'{data_dir}/{fname}') as hdul:
        # print(repr(hdul[2].data))
        all_data.append(hdul[0])
        
        # finding that damn Z header
        if i == 0:
            print(hdul[2].data['Z'])
            # print(repr(hdul[2].header))
            # for head in hdul[2].header:
            #     if head == 'TTYPE64':
            #         print(f'I found {hdul[2].header[head]}!')
            print(repr(hdul[2].header))
            i += 1
            

all_data = np.array(all_data)
```

```python
rows = []

with fits.open("fits/spec-0283-51959-0350.fits") as hdul:
    print(hdul.info())
    for i, hdu in enumerate(hdul):
        for key, value in hdu.header.items():
            rows.append((i, key, value, hdu.header.comments[key]))

header_table = Table(
    rows=rows,
    names=("HDU", "Keyword", "Value", "Comment")
)

header_table
```
